```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Считайте датасет из файла train.csv
df = pd.read_csv('train.csv')
```

```python
# Визуализируйте распределение значений признаков Survived, Pclass, Age, Sex, Parch
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
sns.countplot(x='Survived', data=df)
plt.title('Распределение Survived')

plt.subplot(2, 3, 2)
sns.countplot(x='Pclass', data=df)
plt.title('Распределение Pclass')

plt.subplot(2, 3, 3)
sns.histplot(df['Age'].dropna(), bins=20)
plt.title('Распределение Age')

plt.subplot(2, 3, 4)
sns.countplot(x='Sex', data=df)
plt.title('Распределение Sex')

plt.subplot(2, 3, 5)
sns.countplot(x='Parch', data=df)
plt.title('Распределение Parch')

plt.tight_layout()
plt.show()
```

```python
# Постройте график типа boxplot для столбца Age
plt.figure(figsize=(6, 6))
sns.boxplot(y='Age', data=df)
plt.title('Boxplot для столбца Age')
plt.ylabel('Age')
plt.show()

# Проинтепретируйте полученный график
print("\nBoxplot показывает медиану возраста, а также распределение вокруг нее. Присутствуют выбросы среди старших пассажиров, что видно по точкам за пределами 'усов' графика.")
```

```python
# Постройте график типа pie chart для переменных Survived, Pclass
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart для Survived
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', ax=axes[0], startangle=90, colors=['lightblue', 'lightgreen'])
axes[0].set_title('Процент выживших')
axes[0].set_ylabel('')

# Pie chart для Pclass
df['Pclass'].value_counts().plot.pie(autopct='%1.1f%%', ax=axes[1], startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])
axes[1].set_title('Распределение по Pclass')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Постройте график типа pairplot для всех числовых переменных датасета
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].dropna())
plt.suptitle('Pairplot для числовых переменных', y=1.02)
plt.show()

# Постройте интерактивный sunburst plot для классов и пола
fig = px.sunburst(df, path=['Pclass', 'Sex'], values='PassengerId',
                  color='Pclass', color_continuous_scale='RdBu',
                  title='Sunburst Plot: Пассажиры по классам и полу')
fig.show()
```
